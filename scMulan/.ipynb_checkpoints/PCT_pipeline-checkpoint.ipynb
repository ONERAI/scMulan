### 以/nfs/public/chenyx/scMulan_test/TAA_all.h5ad为例

1. 类叫scMulan，函数是annotate_h5ad()
2. load用户的h5ad数据；
3. 转h5ad，变成uniformed genes

```python
import scanpy as sc
from reference.GeneSymbolUniform.pyGSUni import GeneSymbolUniform
```

## 1. load h5ad
It's recommended that you use h5ad here with raw count (and after your QC)

```python
adata_path = '/nfs/public/chenyx/scMulan_test/TAA_all.h5ad'
```

```python
adata = sc.read_h5ad(adata_path)
```

## 2. transform original h5ad with uniformed genes

```python
adata_GS_uniformed = GeneSymbolUniform(input_adata=adata,
                                 output_dir="/nfs/public/chenyx/scMulan_test/uniformed_adata/",
                                 output_prefix='jmdata_TAA_adata')
```

## 3. process uniformed data (simply norm and log1p)

```python
adata_GS_uniformed
```

```python
sc.pp.normalize_total(adata_GS_uniformed, target_sum=1e4) 
sc.pp.log1p(adata_GS_uniformed)
```

## 4. 输入模型

```python
import scMulan
```

```python
from imp import reload
reload(scMulan)
```

```python
ckp_path = '/nfs/public/cell_gpt_data/dataHub/datasets/datasets/ECA_GO/model_hub/ckpt245000.pt'
```

```python
scml = scMulan.scMulan(ckp_path, adata_GS_uniformed, kv_cache=False) #FIXME remove the print ele, etc
```

```python
adata_pred = scml.predict_cell_types_for_adata()
```

```python
adata_pred.obs.cell_type_from_mulan
```

## 5. visualization

```python
sc.pp.pca(adata_pred)
sc.pl.pca_variance_ratio(adata_pred)
sc.pp.neighbors(adata_pred,n_pcs=10)
sc.tl.umap(adata_pred)
```

```python
sc.pl.embedding(adata_pred,basis='X_umap',color=['cell_type_from_mulan'])
```
